```python
def plot_2d_dataset(data, colors):
    pyplot.figure(figsize(5,5))
    pyplot.scatter(list(map(lambda x: x[0], data[0])), list(map(lambda x: x[1], data[0])), c = data[1], cmap = colors)
```

```python
noisy_circles = datasets.make_circles(noise = 0.05)
plot_2d_dataset(noisy_circles, colors)
```

make_classification

```python
simple_calssification = datasets.make_classification(n_features = 2, n_informative=1, n_redundant=1, n_clusters_per_class=1, random_state=1)
plot_2d_dataset(simple_calssification, colors)
```

```python
calssification = datasets.make_classification(n_features = 2, n_informative=2,n_classes = 4,n_redundant=0, n_clusters_per_class=1, random_state=1)
colors = ListedColormap(["red", "blue", "green", "orange"])
plot_2d_dataset(calssification, colors)
```

```python
from pandas import DataFrame
```

```python
iris_frame = DataFrame(iris.data)
iris_frame.columns = iris.feature_names
iris_frame['target'] = iris.target
```

```python
iris_frame.head()
```

```python
iris_frame.target = iris_frame.target.apply(lambda x : iris.target_names[x])
iris_frame.head()
```

```python
iris_frame[iris_frame.target == 'setosa'].hist('sepal length (cm)')
```

```python
pyplot.figure(figsize(15, 18))

plot_num = 0
for feature_name in iris['feature_names']:
    for target_name in iris['target_names']:
        plot_num += 1
        pyplot.subplot(4,3,plot_num)
        pyplot.hist(iris_frame[iris_frame.target == target_name][feature_name])
        pyplot.title(target_name)
        pyplot.xlabel('cm')
        pyplot.ylabel(feature_name[:-4])
        
```
